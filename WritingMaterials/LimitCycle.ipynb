```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from numpy import array as npa


```

```python
# dx = x - y - x(x**2 + y**2)
# dy = x + y - y(x**2 + y**2)


def dx(x, y, dt):
    return dt*(x - y - x*(x**2 + y**2))
    
    
def dy(x, y, dt):  
    return dt*(x + y - y*(x**2 + y**2))
```

```python
# xy = npa([1.5, 1.5])[None]
xy = npa([0.5, 0.2])[None]
dt = 0.01
for i in range(1, 1000*3):
    x = xy[-1, 0] + dx(xy[-1, 0], xy[-1, 1], dt)
    y = xy[-1, 1] + dy(xy[-1, 0], xy[-1, 1], dt)
#     noise = np.random.randn(2) * dt * 2
    noise = np.random.uniform(-1, 1, 2) * dt * 5
    xy = np.vstack([xy, [x, y]+ noise])
    
plt.plot(xy[:, 0], xy[:, 1])
# plt.axis('equal')
# plt.axis([-2, 2, -2, 2])

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()
plt.gca().set_aspect("equal")
#     plt.axis('equal')
    

# xy
```
