```python
import numpy as np
from numpy import array as npa
from scipy.stats import norm as norm
import scipy.stats as stats

from matplotlib import pyplot as plt 
```

```python
mu = [3,8]
sigma = [1, 0.6]
x = np.linspace(-1, 11, 1000)
y = []
y.append(norm.pdf(x, mu[0], sigma[0]))
y.append(norm.pdf(x, mu[1], sigma[1]))

mu_ = (mu[0]*sigma[1]**2+mu[1]*sigma[0]**2) / (sigma[0]**2+sigma[1]**2)
sigma_ = (sigma[0]**2 * sigma[1]**2)
mu.append( mu_ )
sigma.append( sigma_ )

y.append(norm.pdf(x, mu[2], sigma[2]))


plt.figure()
plt.plot(x, y[0])
plt.plot(x, y[1])
plt.plot(x, y[2])

plt.savefig("BayesianIntegration.svg")


```

# random neuron net

```python
import networkx as nx
```

```python
n = 15
G = nx.fast_gnp_random_graph(n, 0.2, directed=True)
plt.figure(figsize=[5, 5])

cs = np.random.choice(['#0092b3','#F0FAFC'], n)

pos = nx.drawing.layout.circular_layout(G)
nodes = nx.draw_networkx_nodes(G, pos, node_color=list(cs))
nodes.set_edgecolor('k')
edges = nx.draw_networkx_edges(G, pos, edge_color='#a8a8a8', width=2)
nx.write_gexf(G, 'G1.gexf')

plt.savefig("G1.svg")
```
